```python
import altair as alt
from vega_datasets import data

counties = alt.topo_feature(data.us_10m.url, 'counties')
source = data.unemployment.url

def draw_map(cmap):
    m = alt.Chart(counties).mark_geoshape().encode(
        color=alt.Color('rate:Q', scale=alt.Scale(range=cmap.colors)),
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['rate'])
    ).project(
        type='albersUsa'
    ).properties(
        width=300,
        height=220
    )
    m.display()

from cieran import Cieran
cie = Cieran(draw=draw_map)
cie.set_color("#a57263")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Original line with decreased opacity
plt.plot(cie.reward_history, color='skyblue', alpha=0.4)

# Smoothing the data using a rolling mean. Adjust window size as needed.
smoothed = pd.Series(cie.reward_history).rolling(window=289).mean()
plt.plot(smoothed, color='purple', label='Rolling average')  # Using red for the smoothed line, but you can change this
max_val = max(cie.reward_history)
plt.axhline(y=max_val, color='r', linestyle='--', label='Max Value')


plt.legend()
plt.show()
```

```python
from cieran import Cieran
cie = Cieran(draw=draw_map)
cie.set_color("#447ba0")
cie.set_reward_weights([1,0,0,0,0,0,0,0,0])
cie.search()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Set the font to "Helvetica Neue"
plt.rcParams['font.family'] = 'Helvetica Neue'
sns.set_theme(style="whitegrid", palette="pastel")

# Convert reward_history to a DataFrame
df = pd.DataFrame({'Episode Reward': cie.reward_history})

# Smoothing the data using a rolling mean
df['Rolling Average'] = df['Episode Reward'].rolling(window=289).mean()

# Plot using Seaborn
plt.figure(figsize=(6, 6))
sns.lineplot(data=df, linewidth=1)

# Adding a reference line at the maximum value of reward_history
max_val = df['Episode Reward'].max()
plt.axhline(y=max_val, linestyle='--', label='Max Reward')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# SFSO Parties
palette = ['#6268af','#f39f5e','#ea546f','#547d34','#cbd401','#ffff00','#26b300','#792a8f','#9fabd9','#f0da9d','#bebebe']

# Set overall aesthetic
sns.set_style("white", {
    "grid.linestyle": "--"
})

# Set the font to "Futura" or "Helvetica Neue" if Futura isn't available
plt.rcParams['font.family'] = 'Helvetica Neue'

# Convert reward_history to a DataFrame
df = pd.DataFrame({'Episode Reward': cie.reward_history})

# Smoothing the data using a rolling mean
df['Rolling Average'] = df['Episode Reward'].rolling(window=500).mean()

# Plot using Seaborn
plt.figure(figsize=(6, 6))
sns.lineplot(data=df['Episode Reward'], color=palette[0], lw=1, alpha=0.4)

# Adding a reference line at the maximum value of reward_history
max_val = df['Episode Reward'].max()
plt.axhline(y=max_val, color=palette[7], linestyle='--', lw=1.5, label='Max Value Episode')
sns.lineplot(data=df['Rolling Average'], color=palette[7], lw=1, label='Rolling Average (500 Eps.)')


plt.legend(frameon=True, framealpha=1, edgecolor=palette[10], fancybox=False)
plt.tight_layout()
plt.savefig("Q100e0.svg", format="svg")
plt.show()
```
