```python
import altair as alt
from vega_datasets import data

counties = alt.topo_feature(data.us_10m.url, 'counties')
source = data.unemployment.url

def draw_map(cmap):
    m = alt.Chart(counties).mark_geoshape().encode(
        color=alt.Color('rate:Q', scale=alt.Scale(range=cmap.colors)),
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['rate'])
    ).project(
        type='albersUsa'
    ).properties(
        width=300,
        height=220
    )
    m.display()
```

```python
from cieran import Cieran

cie = Cieran(draw=draw_map)
cie.set_color("#7dbcc1")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from matplotlib.colors import ListedColormap

# Assuming the draw function takes a colormap and plots a gradient for demonstration
def draw(cmap):
    gradient = np.linspace(0, 1, 256).reshape(1, -1)
    gradient = np.vstack((gradient, gradient))
    fig, ax = plt.subplots(figsize=(6, 2))
    ax.set_title(cmap.name, fontsize=14)
    ax.imshow(gradient, aspect='auto', cmap=cmap)
    ax.set_axis_off()
    plt.show()
```

```python
# Sample list of ListedColormap objects for demonstration purposes
cmaps = [plt.cm.Reds, plt.cm.Blues, plt.cm.Greens, plt.cm.Purples]

# Create an output widget to display the chart
out = widgets.Output()

# Create a label widget to display the current value
label = widgets.Label(layout=widgets.Layout(margin='0 auto'))

# Create a vertical slider and set its characteristics, but without readout
slider = widgets.IntSlider(
    value=0,
    min=0,
    max=len(cmaps)-1,
    step=1,
    description='',
    disabled=False,
    orientation='vertical',
    readout=False
)

# Function to update the label
def update_label(val):
    label.value = str(len(cmaps) - 1 - val)

# Create a function that will be called on slider value change
def update(val):
    # Interpret the value in reverse
    cmap = cmaps[len(cmaps) - 1 - val]
    with out:
        # Clear the previous plot
        out.clear_output(wait=True)
        # Draw the new plot
        draw(cmap)
    # Update the label
    update_label(val)

# Bind the update function to the slider
slider.observe(lambda change: update(change['new']), names='value')

# Set the initial label value
update_label(slider.value)

# Create a vertical box for the slider and its custom label
slider_with_label = widgets.VBox([label, slider])

# Create a horizontal box layout to display the slider and the chart side by side
layout = widgets.HBox([slider_with_label, out])

# Call the update function once to display the initial chart
update(slider.value)

# Display the layout
layout
```

```python
import csv
from collections import defaultdict

all_data = []
experiment_data = []
with open('all4.csv') as infile:
    reader = csv.reader(infile)
    next(reader)
    for row in reader:
        all_data.append(row)
```

```python
# I want data that looks like:
# columns: [originalId, trialId, epsilon, Q, total_reward1, ...

%load_ext autoreload
%autoreload 2
from cieran import Cieran
import numpy as np
import time

for row in all_data:
#     for params in hyperparams:
    params = hyperparams[0]
    for i in range(10):
        cie = Cieran(draw=draw_map)
        cie.set_color(row[1])
        cie.set_reward_weights([float(w) for w in row[2:11]])
        cie._ranker()

        cie._env.epsilon = params[0]
        cie._env.Q = defaultdict(float)
        cie._env.Q.default_factory = lambda: params[1]
        start_time = time.time()  # Start the timer
        cie.search()
        end_time = time.time()  # End the timer
        
        max_reward = np.dot(cie._search_result.features, cie._env.reward_weights)
        expert_reward = np.dot(cie._ranked_results[0].features, cie._env.reward_weights)

        experiment_data.append([row[0], params[0], params[1],  i + 1, max_reward, max_reward >= expert_reward, cie._search_result.length, end_time - start_time])
```
