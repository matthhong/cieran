```python
%load_ext autoreload
%autoreload 2
from cieran import Cieran
import seaborn as sns
import numpy as np
sns.set_theme(style="white")
rs = np.random.RandomState(50)
import matplotlib.pyplot as plt

x, y = rs.normal(size=(2, 200), scale=2)
def draw_kde(cmap):
    # Set up the matplotlib figure
    fig, ax = plt.subplots()
    
    cmap = plt.get_cmap(cmap)
    sns.kdeplot(x=x, y=y, cmap=cmap, ax=ax, fill=True)
    # Adjust aspect ratio to be equal
    ax.set_aspect('equal')

    ax.set_axis_off()
    ax.set(xlim=(-6, 6), ylim=(-6, 6))
    plt.tight_layout()
    
    plt.show()
    
def draw_heatmap(cmap):
    # Set up the matplotlib figure
    fig, ax = plt.subplots()
    
    cmap = plt.get_cmap(cmap)
    sns.histplot(x=x, y=y, cmap=cmap, ax=ax, stat='density', bins=20)
    # Adjust aspect ratio to be equal
    ax.set_aspect('equal')

    ax.set_axis_off()
    ax.set(xlim=(-6, 6), ylim=(-6, 6))
    plt.tight_layout()

    plt.show()

def draw_hexbin(cmap):
    # Set up the matplotlib figure
    # fig, ax = plt.subplots()
    
    fig, ax = plt.subplots()
    cmap = plt.get_cmap(cmap)
    plt.hexbin(x, y, cmap=cmap, gridsize=20)

    ax.set_aspect('equal')

    ax.set_axis_off()
    ax.set(xlim=(-6, 6), ylim=(-6, 6))
    plt.tight_layout()

    plt.show()
    
def draw_scatter(cmap):
    # Set up the matplotlib figure
    # fig, ax = plt.subplots()
    
    fig, ax = plt.subplots()
    cmap = plt.get_cmap(cmap).reversed()
    plt.scatter(x, y, c='black')

    ax.set_aspect('equal')

    ax.set_axis_off()
    ax.set(xlim=(-6, 6), ylim=(-6, 6))
    plt.tight_layout()

    plt.show()
```
