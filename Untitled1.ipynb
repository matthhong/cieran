```python
import altair as alt
from vega_datasets import data

counties = alt.topo_feature(data.us_10m.url, 'counties')
source = data.unemployment.url

def draw_map(cmap):
    m = alt.Chart(counties).mark_geoshape().encode(
        color=alt.Color('rate:Q', scale=alt.Scale(range=cmap.colors)),
    ).transform_lookup(
        lookup='id',
        from_=alt.LookupData(source, 'id', ['rate'])
    ).project(
        type='albersUsa'
    ).properties(
        width=300,
        height=220
    )
    m.display()

from cieran import Cieran
cie = Cieran(draw=draw_map)
cie.set_color("#447ba0")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

palette = ['#6268af','#f39f5e','#ea546f','#547d34','#cbd401','#ffff00','#26b300','#792a8f','#9fabd9','#f0da9d','#bebebe']

def qplot(reward_history, title, color):

    sns.set_context("paper", rc={"font.size":10,"axes.titlesize":12,"axes.labelsize":8})  

    # Set overall aesthetic
    sns.set_style("white", {
        "grid.linestyle": "--"
    })

    # Set the font to "Futura" or "Helvetica Neue" if Futura isn't available
    plt.rcParams['font.family'] = 'Helvetica Neue'

    # Convert reward_history to a DataFrame
    df = pd.DataFrame({'Episode Reward': [r-10 for r in reward_history]})

    # Smoothing the data using a rolling mean
    df['Rolling Average'] = df['Episode Reward'].rolling(window=500).mean()

    # Plot using Seaborn
    fig, ax = plt.subplots(figsize=(4, 4))  # You might need to adjust the figure size to get the aspect you desire
    sns.lineplot(data=df['Episode Reward'], color=color, lw=1, alpha=0.2, ax=ax)

    # Adding a reference line at the maximum value of reward_history
    max_val = df['Episode Reward'].max()
    ax.axhline(y=max_val, color=color, linestyle='--', lw=1.5, label='Max Value Episode')
    sns.lineplot(data=df['Rolling Average'], color=color, lw=1.5, label='Rolling Average (500 Eps.)', ax=ax)
    ax.text(df.index[0], max_val+0.01, f'{max_val:.2f}', color=color, verticalalignment='bottom', horizontalalignment='left')

    ax.set_ylim(-0.6, 0.6)
#     ax.legend(frameon=True, framealpha=1, edgecolor=palette[10], fancybox=False)
    ax.legend(frameon=True, framealpha=1, edgecolor=palette[10], fancybox=False, loc='upper center', bbox_to_anchor=(0.5, -0.2))


    # Extract the bounding box of the plotting area in figure coordinates
    pos = ax.get_position()

    # Get the width and height of the plotting area
    plot_width = pos.width
    plot_height = pos.height

    # Adjust the figure width and height so that the plotting area is square
    fig_width, fig_height = fig.get_size_inches()
    fig_aspect = fig_width / fig_height
    if plot_width > plot_height:
        fig_width *= plot_height / plot_width
    else:
        fig_height /= plot_width / plot_height
    fig.set_size_inches(fig_width * fig_aspect * (2/3), fig_height * (2/3))

    plt.savefig(title + ".svg", format="svg")
    plt.show()
```
